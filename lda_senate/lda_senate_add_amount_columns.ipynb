```python
import pandas as pd
import numpy as np
import os

folder_path = "lda_senate_output"

# Create empty list to use when bringing in each csv as a df
df_list = []
```

```python
# Read in senate_info CSV containing senator information
senate = pd.read_csv("../senate_info_name_cleaned.csv")
```

```python
# Combine First and Last to create name to look up
senate["First Last"] = senate["First_Actual"] + " " + senate["Last"]
senate
```

```python
# For each csv file in folder...
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        
        # read the CSV file into df
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        
        # For verification to see that loop is working
        print(df.shape)
        
        # add df to list
        df_list.append(df)

        # For each senator...
        total_amounts = []
        for index, row in senate.iterrows():
            print(index)
            df["honoree_name"] = df["honoree_name"].fillna("")
            # search for the full name in the search df
            search_result = df[df['honoree_name'].str.contains(row["First Last"],case=False)]
            # calculate the total amount
            total_amount = search_result['amount'].sum()
            # append the total amount to the list
            total_amounts.append(total_amount)
        
        # Name column based on CSV file and enter amounts
        senate[file_name] = total_amounts
```

```python
# Rename columns
senate = senate.rename(columns={"2013page1-1412.csv":"2013 Contributions",
                                "2014page1-1346.csv":"2014 Contributions",
                                "2015page1-1331.csv":"2015 Contributions",
                                "2016page1-1324.csv":"2016 Contributions",
                                "2017page1-1409.csv":"2017 Contributions",
                                "2018page1-1391.csv":"2018 Contributions",
                                "2019page1-1412.csv":"2019 Contributions",
                                "2020page1-1400.csv":"2020 Contributions",
                                "2021page1-1434.csv":"2021 Contributions",
                                "2022page1-1429.csv":"2022 Contributions",
                                })
```

```python
# Add total column
senate["Total Contributions 2013-2022"] = senate[["2013 Contributions","2014 Contributions",
                                                  "2015 Contributions","2016 Contributions",
                                                  "2017 Contributions","2018 Contributions",
                                                  "2019 Contributions","2020 Contributions",
                                                  "2021 Contributions","2022 Contributions"]].sum(axis=1)
```

```python
# Export
senate.to_csv("../senate_info_with_contributions.csv")
```

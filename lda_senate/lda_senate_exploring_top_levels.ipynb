# This file is for the purpose of exploring the different top level directories of the lda senate database

```python
import requests, json
import pandas as pd

base_url = "https://lda.senate.gov/api/v1/"
```

```python
# Request top-most level and save as table of contents
table_of_contents = requests.get(base_url).json()
print(json.dumps(table_of_contents,indent=4,sort_keys=True))
```

```python
# Produce headers list that contains the different headers from the table of contents 
# (extracted from the keys of the dictionary that was pulled)
headers = list(table_of_contents.keys())
print(headers)
```

# Explore what each header contains

```python
# Header 0 - "Filings"
x=0
response = requests.get("https://lda.senate.gov/api/v1/"+str(headers[x])).json()
print(headers[x])
print(json.dumps(response,indent=4,sort_keys=True))
```
